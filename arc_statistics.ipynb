```python
# ARC Prize statistics
data = []
with open(f"{ROOT_DATA_PATH}/arc-agi_{split}_{inp_or_out}.json","r") as f:
    d = json.load(f)
    data.append(d)

df = pd.DataFrame(data)
df.head()
```

```python
# Creating figure and subplots
TOKEN_LEN=3.0
dfs = []
for i,split in enumerate(["training", "evaluation"]):
    with open(f"{ROOT_DATA_PATH}/arc-agi_{split}_challenges.json","r") as f:
        d_input = json.load(f)
    with open(f"{ROOT_DATA_PATH}/arc-agi_{split}_solutions.json","r") as f:
        d_output = json.load(f)

    inputs = pd.DataFrame([len(str(d_input[key]["train"]) + str(d_input[key]["test"]))/TOKEN_LEN for key in d_input], columns=['length'])
    inputs['type'] = 'challenge'
    outputs = pd.DataFrame([len(str(d_output[key]))/TOKEN_LEN for key in d_output], columns=['length'])
    outputs['type'] = 'solution'
    df = pd.concat([inputs,outputs], axis=0)
    df['split'] = split
    dfs.append(df)
    

stat_df = pd.concat(dfs, axis=0)
sns.barplot(x='split',y='length', hue='type', data=stat_df)
plt.title("Avg Token Length statistics for ARC prize data (1 token ~ 3 characters)")
plt.show()     
        
```
